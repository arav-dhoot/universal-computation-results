```python
import json
import torch
import matplotlib.pyplot as plot
```

```python
with open('bit_xor_data.json', 'r') as f:
  data = json.load(f)

variance_data = dict()
for key in data:
    variance_data[key] = torch.var(torch.tensor(data[key]))
```

```python
for value in variance_data:
  if variance_data[value] != variance_data[value]:
    variance_data[value] = torch.tensor(0.0)
```

```python
variance_data = dict(sorted(variance_data.items(), key=lambda item: item[1]))
layers = variance_data.keys()
values = variance_data.values()
plot.bar(layers, values, color ='maroon',width = 0.5)
plot.plot()
```

```python
print('=' * 50)
print('Values for the Graph (left to right)')
print('=' * 50)
for value in variance_data:
    print(f'{value} => {variance_data[value]}')
print('=' * 50)
```

```python
count = 0
for value in data:
  if len(data[value]) > 0:
    count += 1

print(count)
```

```python
plot_graphs = list()
for value in data:
  if len(data[value]) > 0:
    length = len(data[value])
    x_values = list(range(1, length + 1))
    y_values = data[value]
    plot.title(value)
    plot_graphs.append(plot.plot(x_values[100:], y_values[:100]))
    plot.savefig(f'{value}.png')
    plot.show()
```
